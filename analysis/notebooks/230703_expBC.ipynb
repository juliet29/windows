## Exp A

```python
a00, a01 = h.import_desired_data("A", "15T")

w3a = w.Window_Detect2(a01)
w3a.run_analyis_and_guess(w.make_ewm_smooth, )
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
colors = list(px.colors.qualitative.Plotly)
window_color = colors[0]
```

```python
import plotly.io as pio

colorway = ['#702632', '#A4B494', '#495867', '#912F40',  "#F4442E", "#DB7C26",  "#BB9BB0",]
pio.templates["cardinal"] = go.layout.Template(
    layout=go.Layout(
        colorway=colorway
    ))
# combine templates 
pio.templates.default = 'plotly_white+cardinal'
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.05)

width_num = 1

fig.add_trace(go.Scatter(
    x=w3a.time,
    y=w3a.window_norm,
    legendgroup="window",
    mode='lines',
    line=dict(width=width_num),
    name=r"$W(t)$",
    opacity=0.5,
    marker_color=window_color,
), row = 1, col = 1)

fig.add_trace(go.Scatter(
    x=w3a.time,
    y=w3a.temp_norm, 
    mode='lines',
    line=dict(width=width_num),
    name=r"$T(t)$"
), row = 1, col = 1)

fig.add_trace(go.Scatter(
    x=w3a.time,
    y=w3a.smooth_series, 
    mode='lines',
    line=dict(width=width_num),
    name=r"$\overline{T(t)}$"
), row = 1, col = 1)

fig.add_trace(go.Scatter(
    x=w3a.time,
    y=w3a.dif, 
    mode='lines',
    line=dict(width=width_num),
    name=r"$T'(t)$"
), row = 1, col = 1)

# guesses-----------------
fig.add_trace(go.Scatter(x=w3a.time, y=w3a.window_norm, 
mode='lines',
line=dict(width=width_num),
legendgroup="window",
showlegend=False,
name=r"$W(t)$", 
opacity=0.5,
marker_color=window_color,
), row = 2, col = 1)


fig.add_trace(go.Scatter(x=w3a.time, y=w3a.deriv2,
name=r"$\frac{d^2T'(t)}{dt^2}$", 
mode='lines',
line=dict(width=width_num),
),
row = 2, col = 1)

fig.add_trace(go.Scatter(x=w3a.guess_times, y=w3a.guess_values, name=r"$G(t)$", 
mode='markers'),
row = 2, col = 1)

fig.add_trace(go.Scatter(x=w3a.guess_times, y=w3a.round_guesses, name=r"$G_{round}(t)$", 
mode='markers')
,row = 2, col = 1)

fig.add_trace(go.Scatter(x=w3a.interp_guess.index, y=w3a.interp_guess, 
name=r"$I(t)$", 
mode='lines',
line=dict(width=width_num*1.4, color="black"),)
,row = 2, col = 1)

fig.update_layout(
    margin=dict(l=0,r=0,b=0,t=10),
    width=290, height=300,
    xaxis2= dict(
    tickfont = dict(size=8))
    )

fig.write_image(f"../../summary/230613_buildsys/figs/method_ST.png", scale=4)

fig.show()
```
